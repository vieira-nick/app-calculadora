```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import mean_squared_error, accuracy_score

# Gerando dados fictícios para regressão linear
np.random.seed(42)
X_linear = np.random.rand(100, 1) * 10
y_linear = 2.5 * X_linear + np.random.randn(100, 1) * 2

# Treinamento do modelo de Regressão Linear
X_train, X_test, y_train, y_test = train_test_split(X_linear, y_linear, test_size=0.2, random_state=42)
model_linear = LinearRegression()
model_linear.fit(X_train, y_train)
y_pred = model_linear.predict(X_test)
print(f"Coeficiente: {model_linear.coef_[0][0]}, Intercepto: {model_linear.intercept_[0]}")
print(f"Erro Quadrático Médio: {mean_squared_error(y_test, y_pred)}")

# Gerando dados fictícios para regressão logística
X_logistic = np.random.rand(100, 1) * 10
y_logistic = (X_logistic > 5).astype(int).ravel()

# Treinamento do modelo de Regressão Logística
X_train, X_test, y_train, y_test = train_test_split(X_logistic, y_logistic, test_size=0.2, random_state=42)
model_logistic = LogisticRegression()
model_logistic.fit(X_train, y_train)
y_pred = model_logistic.predict(X_test)
print(f"Acurácia Regressão Logística: {accuracy_score(y_test, y_pred)}")

# Gerando dados fictícios para K-Means
X_kmeans = np.vstack([np.random.randn(50, 2) + np.array([2, 2]), np.random.randn(50, 2) + np.array([6, 6])])
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_kmeans)
labels = kmeans.labels_
plt.scatter(X_kmeans[:, 0], X_kmeans[:, 1], c=labels, cmap='viridis')
plt.title("Clusters K-Means")
plt.show()

# Gerando dados fictícios para árvore de decisão
X_tree = np.random.rand(100, 2) * 10
y_tree = (X_tree[:, 0] + X_tree[:, 1] > 10).astype(int)
X_train, X_test, y_train, y_test = train_test_split(X_tree, y_tree, test_size=0.2, random_state=42)
model_tree = DecisionTreeClassifier(max_depth=3)
model_tree.fit(X_train, y_train)
y_pred = model_tree.predict(X_test)
print(f"Acurácia Árvore de Decisão: {accuracy_score(y_test, y_pred)}")

# Visualizando a árvore de decisão
plt.figure(figsize=(10, 6))
plot_tree(model_tree, filled=True, feature_names=['Feature 1', 'Feature 2'])
plt.show()
```
